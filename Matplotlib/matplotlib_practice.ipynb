### Matplotlib example Workflow

```python
#0. import matplotlib and get ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

#3. Plot data
ax.plot(x,y)

#4. customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#5. Save & show
fig.savefig('../Matplotlib/images/sample-plot.png') 
```

## Making figures with NumPy Arrays

We want:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
#Create data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
#Use same data to make a scatterplot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
#Another Scatterplot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(x=nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
       ylabel='Price ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots


```python
#subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
#plot to each
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```
